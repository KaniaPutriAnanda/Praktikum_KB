## Import Library

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## Membuka Dataset dan Menampilkan Data Teratas

```python
df = pd.read_csv("student-mat.csv")
df.head()
```

## Grafik Batang

```python
X = list(df["age"])
Y = list(df["absences"])

plt.figure(figsize=(10,10))
plt.title("Survey Siswa di Kursus Matematika")
plt.xlabel("Ketidak hadiran")
plt.ylabel("Umur Siswa")
plt.bar(X,Y, color = "purple")
plt.show()
```

## Kesimpulan:
### Pada grafik,siswa dengan umur 18 tahun memiliki jumlah ketidak hadiran paling banyak yaitu lebih dari 70 kali ketidak hadiran, siswa dengan umur 17 dan 16 tahun memiliki jumlah ketidak hadiran lebih dari 50 kali, siswa dengan umur 19 tahun memiliki jumlah ketidak hadiran 40 kali, siswa dengan umur 15 tahun memiliki jumlah ketidak hadiran lebih dari 20 kali, siswa dengan umur 22 tahun memiliki jumlah ketidak hadiran lebih dari 10 kali, dan siswa dengan umur 21 tahun memiliki jumlah ketidak hadiran paling sedikit yaitu kurang dari 10 kali ketidak hadiran.

## Grafik Pie

```python
data_pie = df['Fjob'].value_counts().rename('Fjob').reset_index(name = 'id_count')
data_pie.head()
```

```python
plt.pie(data_pie.id_count,
        labels = data_pie['index'],
        startangle = 180,
        autopct = '%.1f%%',
        explode = (0.05,0,0,0,0),
        shadow = True)
plt.title('Pekerjaan Ayah')
plt.show()
```

## Kesimpulan:
### Pekerjaan ayah dari siswa yang disurvey kebanyakan bekerja sebagai lainnya yaitu dengan persentase 54.9%, yang artinya jenis pekerjaan tersebut bukan dari 4 kategori yang tersedia. Kategori terbanyak kedua yaitu ayah yang bekerja sebagai pelayan administrasi/polisi yaitu sebanyak 28.1%. Sedangkan ayah yang bekerja sebagai guru, tidak bekerja dan tenaga medis nilai persentasenya secara berurutan yaitu 7.3%, 5.1% dan 4.6%

## Grafik Scatterplot

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10,5))
sns.scatterplot(
    data = df,
    hue = 'sex',
    x = 'absences',
    y = 'age')
plt.title('Grafik Perbandingan Pekerjaan Orang Tua Siswa')
plt.show()
```

## Penjelasan:
### Ketidak hadiran paling banyak adalah siswa laki-laki di usia 18 tahun. Kemudian pada grafik dapat dilihat bahwa siswa laki-laki mendominasi ketidak hadiran dibanding siswa perempuan.
